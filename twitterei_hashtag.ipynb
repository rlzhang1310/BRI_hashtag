```python
import os
import logging
import boto3
import zipfile
import json
import io
import pandas as pd
import ast
import math
import matplotlib.pyplot as plt
from io import BytesIO
```

```python
s3_host = 'https://obj.umiacs.umd.edu'
access_key_id = "xxxxx"
secret_access_key = "xxxxx"

s3 = boto3.client('s3', 
                  endpoint_url=s3_host, 
                  aws_access_key_id=access_key_id, 
                  aws_secret_access_key=secret_access_key)

bucket = 'twitter.ei'
```

```python
def process(file):
    zip_object = s3.get_object(Bucket=bucket, Key=file)
    zip_contents = zip_object['Body'].read()
    zip_file = zipfile.ZipFile(io.BytesIO(zip_contents), 'r')
    for file_info in zip_file.infolist():
        with zip_file.open(file_info) as csv_file:
            df = None
            try:
                if df == None:
                    df = pd.read_csv(csv_file)
                else:
                    df = pd.concat([df, pd.read_csv(csv_file)], axis=0)
            except:
                print(f"{csv_file.filename} is not a csv file")
    return df
```

```python
def helper(row, hashtag_dict):
    hashtags = row['hashtags']
    if (type(hashtags) == str and hashtags == 'nan') or (type(hashtags) == float and math.isnan(row['hashtags'])):
        return
    hashtags = ast.literal_eval(hashtags)
    if type(hashtags) != list:
        print(hashtags)
        print(type(hashtags))
        hashtags = [hashtags]
    for h in hashtags:
        # if type(hashtags)
        if h in hashtag_dict:
            hashtag_dict[h] += 1
        else:
            hashtag_dict[h] = 1
    return
```

```python
folder_prefix_lst = ["2019_08/china_082019_1/china_082019_1_tweets_csv_unhashed.zip",
                     "2019_08/china_082019_2/china_082019_2_tweets_csv_unhashed.zip",
                     "2019_08/china_082019_3/china_082019_3_tweets_csv_unhashed.zip",
                     "2020_05/china_052020/china_052020_tweets_csv_unhashed.zip", 
                     "2020_09/thailand_092020/thailand_092020_tweets_csv_unhashed.zip"
                     ]

hashtag_lst = []
# pd.set_option('display.float_format', lambda x: '%.0f' % x)

for folder in folder_prefix_lst:
    print(folder)
    data = process(folder)
    hashtag_dict = dict()
    data.apply(lambda row: helper(row, hashtag_dict), axis = 1)
    print(hashtag_dict)
    hashtag_lst.append(hashtag_dict)
```

```python
# dict to df
df_lst = []
for d in hashtag_lst:
    hashtag_df = pd.DataFrame(list(d.items()), columns=['hashtag', 'count'])
    df_lst.append(hashtag_df)
```

```python
for i in range(5):
    df_lst[i] = df_lst[i].sort_values(by='count', ascending=False)
    print(df_lst[i])
```

```python
## downloads data
title = ["china_082019_1", "china_082019_2", "china_082019_3", "china_052020", "thailand_092020"]
i = 0
for df in df_lst:
    df.to_csv(f"~/Coding/buntain/hashtag_data/twitterrei_{title[i]}.csv", index=False)
    i += 1
```
